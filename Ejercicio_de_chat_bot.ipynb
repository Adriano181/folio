```python
# Situación: Tu amigo te pide prestado dinero por quinta vez y aún no te ha devuelto lo anterior

from gtts import gTTS
import IPython.display as ipd

def chatbot(modo):
    if modo == "agresivo":
        respuesta = (
            "¡Otra vez pidiéndome plata! Ya me debes varias veces y no has pagado nada. "
            "¡Olvídalo, no te voy a prestar ni un peso más!"
        )

    elif modo == "pasivo":
        respuesta = (
            "Bueno... supongo que podría ayudarte otra vez, aunque todavía no me has pagado lo anterior. "
            "Está bien, pero me preocupa un poco."
        )

    elif modo == "asertivo":
        respuesta = (
            "Te aprecio mucho, pero no puedo prestarte nuevamente porque aún no me has devuelto "
            "lo que te presté antes. Cuando nos pongamos al día, con gusto lo consideramos."
        )

    else:
        respuesta = "Modo no reconocido. Elige: agresivo, pasivo o asertivo."

    return respuesta

modo = input("Elige el modo de comunicación (agresivo, pasivo, asertivo): ").lower()
respuesta = chatbot(modo)

print("\n Chatbot:", respuesta)


tts = gTTS(respuesta, lang='es')
tts.save("respuesta.mp3")
ipd.display(ipd.Audio("respuesta.mp3"))
```

```python
def analizar_situacion(texto):
    """
    Analiza un texto para identificar si la situación descrita es agresiva, pasiva o asertiva.
    """
    # Convertir el texto a minúsculas para un análisis sin distinción de mayúsculas y minúsculas
    texto = texto.lower()

    # Palabras clave para cada tipo de comportamiento
    palabras_clave_agresivo = ["grito", "gritó", "insulto", "insultó", "pelea", "amenazó", "ordenó", "exijo", "exigió", "enfureció"]
    palabras_clave_pasivo = ["callé", "calla", "evité", "evitó", "miedo", "no pude decir", "cedí", "dejé pasar", "silencio", "sufro"]
    palabras_clave_asertivo = ["expresé", "expresó", "dije", "digo", "escuché", "acuerdo", "respeto", "negocié", "negoció", "opinión"]

    # Contar las coincidencias de palabras clave en el texto
    conteo_agresivo = sum(1 for palabra in palabras_clave_agresivo if palabra in texto)
    conteo_pasivo = sum(1 for palabra in palabras_clave_pasivo if palabra in texto)
    conteo_asertivo = sum(1 for palabra in palabras_clave_asertivo if palabra in texto)

    # Diccionarios de respuestas
    respuestas_agresivas = [
        "Parece que esa situación refleja un comportamiento agresivo. En un comportamiento agresivo, se prioriza la propia necesidad sin respetar los derechos o sentimientos de los demás. A menudo se manifiesta con gritos, insultos o amenazas.",
        "El comportamiento que describes suena a agresividad. La agresión se caracteriza por imponer la voluntad de uno sobre los demás, a veces a expensas de la dignidad ajena. Es importante encontrar maneras más saludables de comunicarse.",
        "Eso es un claro ejemplo de un comportamiento agresivo. Las personas agresivas suelen invadir el espacio personal y emocional de otros para conseguir lo que quieren."
    ]
    respuestas_pasivas = [
        "Esa situación parece ser un ejemplo de comportamiento pasivo. Una persona con comportamiento pasivo tiende a evitar el conflicto a toda costa, lo que puede llevar a que sus propias necesidades no sean satisfechas y a que se sienta menospreciada.",
        "Lo que describes es típico de un comportamiento pasivo. El pasivo se disculpa en exceso y evita expresar sus opiniones por temor al rechazo o al enojo. Aprender a decir 'no' de manera respetuosa es un buen paso.",
        "Ese tipo de interacción es característica de un comportamiento pasivo. La pasividad puede generar frustración y resentimiento, ya que la persona no se defiende ni se posiciona."
    ]
    respuestas_asertivas = [
        "¡Excelente! Lo que describes es un comportamiento asertivo. La asertividad implica expresar tus opiniones, sentimientos y derechos de manera honesta y respetuosa, sin agredir ni someterse a los demás.",
        "Esa es una forma de comunicación asertiva. En la asertividad, la persona busca un equilibrio, respetando a los demás mientras se respeta a sí misma. Es la forma más sana de resolver conflictos.",
        "¡Qué buen ejemplo de asertividad! Se trata de saber negociar y expresar lo que sientes sin pisar a nadie. Es una habilidad muy valiosa para una sana convivencia."
    ]

    # Determinar el tipo de comportamiento
    if conteo_agresivo > conteo_pasivo and conteo_agresivo > conteo_asertivo:
        import random
        return random.choice(respuestas_agresivas)
    elif conteo_pasivo > conteo_agresivo and conteo_pasivo > conteo_asertivo:
        import random
        return random.choice(respuestas_pasivas)
    elif conteo_asertivo > conteo_agresivo and conteo_asertivo > conteo_pasivo:
        import random
        return random.choice(respuestas_asertivas)
    else:
        # En caso de no haber suficientes palabras clave o una mezcla
        return "Hmmm, es difícil de clasificar. Podrías darme más detalles sobre cómo reaccionaste o cómo se sintió la situación?"

# Bucle principal del chatbot
def chatbot():
    print("¡Hola! Soy un chatbot que puede ayudarte a entender diferentes tipos de comportamientos.")
    print("Describe una situación en la que hayas interactuado con alguien. Para salir, escribe 'salir'.")

    while True:
        situacion = input("\n> Describe tu situación: ")
        if situacion.lower() == "salir":
            print("¡Adiós! Que tengas un buen día.")
            break

        respuesta = analizar_situacion(situacion)
        print(f"\nChatbot: {respuesta}")

# Iniciar el chatbot
if __name__ == "__main__":
    chatbot()
```

```python
from gtts import gTTS
import IPython.display as ipd

def chatbot(modo):
    """
    Define y devuelve la respuesta del chatbot según el modo de comunicación.
    La puntuación se usa para mejorar la entonación y el realismo de la voz.
    """
    if modo == "agresivo":
        respuesta = (
            "¡Otra vez! ¡No puede ser! Ya me debes un montón de veces... y todavía no has pagado nada. "
            "¡Olvídalo, no te voy a prestar ni un solo peso más!"
        )

    elif modo == "pasivo":
        respuesta = (
            "Ay... no sé... supongo que podría ayudarte otra vez, ¿no?... Aunque todavía no me has pagado lo anterior... "
            "Está bien... pero me preocupa un poco, ¿sabes?..."
        )

    elif modo == "asertivo":
        respuesta = (
            "Te aprecio mucho como amigo, pero en este momento no puedo prestarte dinero nuevamente. "
            "Aún no me has devuelto lo que te presté antes, y eso me pone en una situación difícil. "
            "Cuando nos pongamos al día con lo anterior, con gusto lo consideramos de nuevo. Gracias por entender."
        )

    else:
        respuesta = "Modo no reconocido. Por favor, elige entre: agresivo, pasivo o asertivo."

    return respuesta

# --- Ejecución del programa ---

modo = input("Elige el modo de comunicación (agresivo, pasivo, asertivo): ").lower()
respuesta = chatbot(modo)

print("\nChatbot:", respuesta)

# Generar y reproducir el audio
try:
    tts = gTTS(text=respuesta, lang='es', slow=False)
    tts.save("respuesta.mp3")
    print("Reproduciendo audio...")
    ipd.display(ipd.Audio("respuesta.mp3"))
except Exception as e:
    print(f"Ocurrió un error al generar o reproducir el audio: {e}")
    print("Asegúrate de tener instaladas las bibliotecas gTTS y IPython.display, y que tu entorno de ejecución lo permita.")
```
